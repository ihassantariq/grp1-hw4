```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import csv
import numpy as np
from sklearn.cluster import KMeans
```

```python
def page_scrapping(web_link):
    
    
    source = requests.get(web_link).text
    
    soup = BeautifulSoup (source, 'lxml')
    
    info_matrix = []
    
    for announcement in soup.find_all('div', class_ = 'listing-item_body'):
        
        try:
            title = announcement.a.text
        
            price = announcement.li.text
        
            v=[]
            i=0
        
            for dati in announcement.find_all('div', class_ = 'lif__data'):
                a = dati.text
                v.append(a)
                i = i+1
            
            locali = v[0]
            superficie = v[1]
            bagni = v[2]
            piano = v[3]
        
            link =announcement.a['href']
            if link.find('https://www.immobiliare.it') == -1:
                link = 'https://www.immobiliare.it' + link
         
            source_desc = requests.get(link).text
            soup1 = BeautifulSoup (source_desc, 'lxml')
            description = soup1.find('div', class_ = 'description-text').text
       
            announce = [title, price, locali, superficie, bagni, piano, link, description]
        
            info_matrix.append(announce)
        except:
            continue
          
    return info_matrix
    
```

```python
A = page_scrapping('https://www.immobiliare.it/vendita-case/roma/?criterio=rilevanza&pag=1')
matrix = np.array([np.array(xi) for xi in A])

for i in range(2, 450):
    A = page_scrapping('https://www.immobiliare.it/vendita-case/roma/?criterio=rilevanza&pag=' + str(i))
    newMatrix = np.array([np.array(xi) for xi in A])
    print("Page data downloaded: " + str(i))
    matrix = np.concatenate((matrix, newMatrix), axis = 0)
```

```python
sub_data = data.iloc[1:10,]
```

```python
sub_data
```

```python
for i in range(1, len(sub_data)+1):
    
    sub_data["Price"][i] = sub_data["Price"][i].replace(".", "")
    sub_data["Price"][i] = float(sub_data["Price"][i])
    
    sub_data["Piano"][i] = sub_data["Piano"][i].replace(" \n", "")
    sub_data["Piano"][i] = sub_data["Piano"][i].replace("\xa0   ", "")
    
    if sub_data["Piano"][i] == 'T':
        sub_data["Piano"][i] = 0
        
    if sub_data["Piano"][i] == 'R':
        sub_data["Piano"][i] = 0.5
    
    if sub_data["Piano"][i] == "A":
        sub_data["Piano"][i] = 100
        
    sub_data["Piano"][i] = float(sub_data["Piano"][i])
    
    sub_data["Locali"][i] = sub_data["Locali"][i].replace("+", "")
    sub_data["Bagni"][i] = sub_data["Bagni"][i].replace("+", "")
        
```

```python
sub_data
```

```python
kmeans = KMeans(n_clusters=3).fit(sub_data)
```

```python
kmeans.labels_
```

```python
cls = []
```

```python
cls = dict()

for i in range(1, len(kmeans.labels_)):
    if kmeans.labels_[i] not in cls:
        cls[kmeans.labels_[i]] = [i]
    elif kmeans.labels_[i] in cls:
        cls[kmeans.labels_[i]].append(i)
        
cls
```
