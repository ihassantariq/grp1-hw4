```python
import pickle
import pandas as pd
import re
import numpy as np
```

```python
df_info = pd.read_pickle("df_info")
```

```python
df_tfidf = pd.read_pickle("df_tfidf")
```

```python
from sklearn.cluster import KMeans
```

## Cluster for info

```python
kmeans_info = KMeans(n_clusters=10).fit(df_info)
```

```python
cls_info = dict()

for i in range(1, len(kmeans_info.labels_)):
    if kmeans_info.labels_[i] not in cls_info:
        cls_info[kmeans_info.labels_[i]] = [i]
    elif kmeans_info.labels_[i] in cls_info:
        cls_info[kmeans_info.labels_[i]].append(i)
        
cls_info
```

## Cluster for description

```python
kmeans_desc = KMeans(n_clusters=10).fit(df_tfidf)
```

```python
cls_desc = dict()

for i in range(1, len(kmeans_desc.labels_)):
    if kmeans_desc.labels_[i] not in cls_desc:
        cls_desc[kmeans_desc.labels_[i]] = [i]
    elif kmeans_desc.labels_[i] in cls_desc:
        cls_desc[kmeans_desc.labels_[i]].append(i)
        
cls_desc
```

```python
l_desc = []
for i in range(10):
    l_desc.append(len(set(cls_desc.get(i))))
    
l_info = []
for i in range(10):
    l_info.append(len(set(cls_info.get(i))))
    
```

Info on the columns, desc on the raw.
Symmetric matrix: calculate only half

```python
mat_inters = np.zeros((10,10))
for i in range(10):
    for j in range(10):
        if i == j:
            mat_inters = 1
        else:
            while i < j:
                mat_inters[i][j] = len(set(cls_desc.get(i)).intersection(cls_info.get(j)))
```

jaccard_similarity

```python
mat_jac = np.zeros((10,10))
```

```python
for i in range(10):
    for j in range(10):
        if i == j:
            mat_jac = 1
        else:
            while i < j:
                mat_jac[i][j] = mat_inters[i][j] / (l_info[j] + l_desc[i] - mat_inters[i][j])
```
